# Shah Jo Risalo Dataset Formatter

A minimal notebook for loading the raw CSV, cleaning the rows we need, and exporting a compact JSON structure for downstream use.

## Dataset at a glance

- Source: `base_dataset/risalo.csv` from the [AMBILE Bhittaipedia project](https://bhittaipedia.org/sur-kalyan/d-1/1).
- Goal: convert the tabular dataset into a clean JSON export
- Kept fields: melody, chapter, type, roman text, explanation, and English translation

```python
import pandas as pd
import json

df = pd.read_csv("base_dataset/risalo.csv")
df.head()
```

Inspect the distinct values for melody, chapter, and type before moving into cleaning and export.

```python
unique_summary = {
    "Melody": df["Melody"].dropna().unique(),
    "Chapter": df["Chapter"].dropna().unique(),
    "Type": df["Type"].dropna().unique(),
}

for label, values in unique_summary.items():
    print(f"{label} ({len(values)} unique values)")
    print(values)
    print()
```

## Cleaning plan

Keep only the fields needed for the final JSON and remove rows that do not have Roman Script.

```python
TO_DROP = [
    "Row_ID",
    "Melody Number",
    "Chapter Number",
    "Bait / Vaayi Number",
    "Sindhi Arabic Perso",
    "Devanagri Script",
    "Urdu Translation ( Agha Saleem )",
    "Punjabi Translation ( Kartar Singh Arsh )",
]

# Drop the columns that are not needed
df = df.drop(columns=TO_DROP)
```

```python
df.info()
```

```python
print("Count of rows where Roman Script is null:", df["Roman Script"].isnull().sum())
```

```python
# Rows where Roman Script is null
df[df["Roman Script"].isnull()]
```

```python
# Delete rows where Roman Script is null
df = df.dropna(subset=["Roman Script"])
df.info()
```

```python
# Clean all the columns by stripping leading and trailing whitespace
for col in df.columns:
    df[col] = df[col].str.strip()
```

```python
# See for duplicates
print("Count of duplicate rows:", df.duplicated().sum())
```

```python
# Conver NaN values to empty strings
df = df.fillna("")
```

## Final export

The notebook finishes by turning the cleaned dataframe into a compact list of dictionaries and writing it to `formatted_dataset/risalo.json`.

```python
# Convert the dataframe to a list of dictionaries
records = df.to_dict(orient="records")
# Create the final list of dictionaries in the desired format
final_data = []
for index, record in enumerate(records):
    final_data.append(
        {
            "id": index + 1,
            "metadata": {
                "melody": record["Melody"],
                "chapter": record["Chapter"],
                "type": record["Type"],
            },
            "content": {
                "roman_text": record["Roman Script"],
                "explanation": record["Explanation"],
                "english_translation": record["English Translation ( Amar Fayaz Buriro )"],
            },
            "searchable_text": f"{record['Type']} {record['Chapter']} {record['Melody']} | {record['Roman Script']} | {record['Explanation']} | {record['English Translation ( Amar Fayaz Buriro )']}",
        }
    )
    
# Save the final data to a JSON file
with open("formatted_dataset/risalo.json", "w", encoding="utf-8") as f:
    json.dump(final_data, f, ensure_ascii=False, indent=4)
```

## Data quality check

Analyze completeness of the three key text fields: roman script, explanation, and English translation.

```python
# Analytics for text-field completeness

roman_col = "Roman Script"
explanation_col = "Explanation"
translation_col = "English Translation ( Amar Fayaz Buriro )"

# Treat non-empty, non-whitespace strings as present
presence = pd.DataFrame({
    "roman": df[roman_col].astype(str).str.strip().ne(""),
    "explanation": df[explanation_col].astype(str).str.strip().ne(""),
    "translation": df[translation_col].astype(str).str.strip().ne(""),
})

total_rows = len(df)

# Summary counts for all combinations of presence/absence
combo_counts = (
    presence.value_counts()
    .rename("count")
    .reset_index()
    .sort_values(["roman", "explanation", "translation"], ascending=[False, False, False])
)

combo_counts["percentage"] = (combo_counts["count"] / total_rows * 100).round(2)
combo_counts["pattern"] = combo_counts.apply(
    lambda r: "+".join(
        [name for name in ["roman", "explanation", "translation"] if r[name]]
    ) if any(r[["roman", "explanation", "translation"]]) else "none",
    axis=1,
)

display(combo_counts[["pattern", "roman", "explanation", "translation", "count", "percentage"]])

# Refined headline analytics
analytics = pd.DataFrame(
    {
        "metric": [
            "Rows with roman + explanation + translation",
            "Rows with roman + explanation only",
            "Rows with roman + translation only",
            "Rows with explanation + translation only",
            "Rows with roman only",
            "Rows with explanation only",
            "Rows with translation only",
            "Rows with none of the three",
        ],
        "count": [
            int((presence["roman"] & presence["explanation"] & presence["translation"]).sum()),
            int((presence["roman"] & presence["explanation"] & ~presence["translation"]).sum()),
            int((presence["roman"] & ~presence["explanation"] & presence["translation"]).sum()),
            int((~presence["roman"] & presence["explanation"] & presence["translation"]).sum()),
            int((presence["roman"] & ~presence["explanation"] & ~presence["translation"]).sum()),
            int((~presence["roman"] & presence["explanation"] & ~presence["translation"]).sum()),
            int((~presence["roman"] & ~presence["explanation"] & presence["translation"]).sum()),
            int((~presence["roman"] & ~presence["explanation"] & ~presence["translation"]).sum()),
        ],
    }
)

analytics["percentage"] = (analytics["count"] / total_rows * 100).round(2)

display(analytics)

print(f"Total rows: {total_rows}")
```
